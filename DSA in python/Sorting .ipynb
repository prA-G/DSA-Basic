# Sorting

Sorting is the process of arranging data in a particular order, usually ascending (small ‚Üí large) or descending (large ‚Üí small).

It‚Äôs like arranging your books on a shelf from shortest to tallest or lining up your friends from youngest to oldest. üìöüë´

## Why Sorting is Important?

- Helps in searching efficiently (like binary search). üîç

- Makes data organized and readable.

- Essential for DSA problems like finding pairs, duplicates, or max/min efficiently.

## Sorting Table

| Sorting Method      | Concept                                                                              | Pythonic Example              |

| **Bubble Sort**     | Compare adjacent elements and swap if out of order; repeat.                          | See below                     |

## Bubble Sort 

Compare adjacent elements, swap if they‚Äôre in the wrong order, and keep ‚Äúbubbling‚Äù the biggest ones to the end with each pass.

**Key facts**

- Stable (equal items keep order)

- In-place (O(1) extra space)

- Best case O(n) if already sorted (with early-exit check)

- Worst/Average O(n¬≤)

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]  # Swap
    return arr

nums = [5, 2, 9, 1, 5]
print(bubble_sort(nums))  # Output: [1, 2, 5, 5, 9]
```

Intuition: Big numbers ‚Äúbubble up‚Äù to the end. üõÅüí®

## Selection Sort

Repeatedly pick the minimum (or maximum) from the unsorted portion and place it at the next position in the sorted portion.
Think: pick the smallest book from the pile and put it at the leftmost empty spot ‚Äî rinse & repeat.

**Properties**

- In-place: O(1) extra space

- Not stable by default (equal items can change order) ‚Äî but you can implement a stable variant.

- Time: O(n¬≤) always (doesn‚Äôt get faster if array is nearly sorted)

- Use-case: educational, tiny arrays, or when memory is super tight.

```python
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

nums = [64, 25, 12, 22, 11]
print(selection_sort(nums))  # Output: [11, 12, 22, 25, 64]
```

Intuition: Pick the smallest remaining element and put it in the correct place. üèÜ

## Insertion Sort

Insertion Sort is a simple, comparison-based sorting algorithm where elements are picked one by one from the unsorted portion and inserted into their correct position in the sorted portion of the array (just like sorting cards in your hand).

**Properties**

- Stable, in-place, adaptive (fast on nearly-sorted arrays).

- Best: O(n) (already sorted with minimal checks) ‚Äî with swaps/shifts O(n) each step in worst case.

- Average/Worst: O(n¬≤).

- Space: O(1).

```python
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i-1
        while j >= 0 and arr[j] > key:
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = key
    return arr

nums = [12, 11, 13, 5, 6]
print(insertion_sort(nums))  # Output: [5, 6, 11, 12, 13]
```

Intuition: Take one element at a time and insert it into the right place among sorted elements. ‚ú®

## Pythonic Way

```python
nums = [3, 1, 4, 1, 5, 8]
sorted_nums = sorted(nums)  # Returns a new sorted list
nums.sort()                  # Sorts the list in-place
print(sorted_nums)  # [1, 1, 3, 4, 5, 9]
print(nums)         # [1, 1, 3, 4, 5, 9]
```

Intuition: Python handles it all for you with TimSort ‚Äî efficient and elegant. üíñ
